# Venda por Ano

Criar uma coluna chamada "Ano", para poder agrupar as vendas posteriormente 

```python
base["Ano"] = base["Order Date"].dt.year
```

```python
base.head()
```

Agrupar as vendas por ano

```python
base.groupby(["Ano"]).sum()
```

```python
base.groupby("Ano")["Sales"].sum().plot.bar();
```

Importando o numpy e melhorando o visual

```python
import numpy as np
import matplotlib.pyplot as plt
```

Atribuir os valores de X como colunas (anos), e Y como valores

Ajustar as pontualidades graficas

```python
soma_ano = base.groupby("Ano")["Sales"].sum()

fig, ax = plt.subplots(figsize = (10, 8))

#Barras
ax.bar(soma_ano.index,
       soma_ano.values,
       linewidth=2,
       color='#8c005b'
             )
#Titulo
ax.set_title("Vendas por ano",
             fontsize=18,
             fontweight='bold',
             color='#d8008d'
            )
#Rotulos de dados - Arrumando os rotulos
for i in range(0,4):
    ax.annotate(soma_ano.values[i],   #texto|Rotulo
                (soma_ano.index[i],   #Posição x
                soma_ano.values[i] ),#Posição y
                ha="center")

#Legenda do eixo X (barras)
ax.xaxis.set_ticks([2015, 2016, 2017, 2018])
ax.tick_params(axis="x",labelsize=14 )

# ax.annotate()

plt.show()
```

# Vendas por mês

```python
base["Mes"] = base["Order Date"].dt.month
```

```python
base.head()
```

Agrupando por mês e por ano para que fique mais visual

```python
soma_mes = base.groupby(["Mes", "Ano"])["Sales"].sum()
```

```python
soma_mes.head()
```

Utilizar o reset.index() para transformar o indice em coluna

```python
soma_mes = soma_mes.reset_index()
```
